```python

# 导入相关模块
import numpy as np
import pandas as pd
```

```python
# 查看数据,数据清洗

import os
# os.chdir(r'C:\Users\Administrator\Desktop')
# 创建工作路径

df = pd.read_excel('2018 “建行杯”互联网+大学生创新创业大赛 全国报名项目统计（数据来自大创网，共约50万项目数据）.xlsx', skiprows=1)

df = df.dropna()
print('初步清洗后数据量为%i条' % len(df))
# 读取数据
df.iloc[0]
```

```python
# 省份去除一些词，匹配下面的地图绘制
df["project_province"] = df["project_province"].str.replace("省|市|自治区|壮族|回族|生产建设兵团|维吾尔|特别行政区", "")
df.head(1)
```

```python
# 项目名称统计
data = df['project_name'].tolist()
print(len(data))
print(data[0])
```

```python
# 已获得投资的领域
df_investment_industry = df[df["Investment"]=="已获投资"]["project_industry"].value_counts()
print(type(df_investment_industry))
# df_investment_industry.index.tolist()
# df_investment_industry.values.tolist()
df_investment_industry.sort_values(ascending=False)[:10]
```

```python
# 已获得投资的福建省高校
df_investment_school2 = df[(df["Investment"]=="已获投资") & (df["project_province"]=="福建")]["project_school"].value_counts()
print(type(df_investment_school2))
df_investment_school2.sort_values(ascending=False)[:10]
```

```python
# 已获得投资的高校
df_investment_school = df[df["Investment"]=="已获投资"]["project_school"].value_counts()
print(type(df_investment_school))
# df_investment_school.index.tolist()
# df_investment_school.values.tolist()
df_investment_school.sort_values(ascending=False)[:10]
```

```python
# 项目领域统计
df_industry = df["project_industry"].value_counts()
print(type(df_industry))
# df_industry.index.tolist()
# df_industry.values.tolist()
df_industry.sort_values(ascending=False)[:10]
```

```python
# 高校统计
df_school = df["project_school"].value_counts()
print(type(df_school))
# df_school.index.tolist()
# df_school.values.tolist()
df_school.sort_values(ascending=False)[:10]
```

```python
# 项目省份统计
df_province = df["project_province"].value_counts()
print(type(df_province))
# df_province.index.tolist()
# df_province.values.tolist()
df_province.sort_values(ascending=False)[:10]
```

```python
from wordcloud import WordCloud
import jieba
import PIL.Image as Image
import re

#  中文结巴分词
def chinese_jieba(text):
    wordlist_jieba=jieba.cut(text)
    space_wordlist=" ".join(wordlist_jieba)
    
    # 匹配中文
    pat=re.compile(r'[\u4e00-\u9fa5]+')
    result=pat.findall(space_wordlist)
    sentence_depart = ' '.join(result)

    return sentence_depart

words = chinese_jieba(''.join(data2))
print(words[:50])
```

```python
# 制作词云
# mask_pic = np.array(Image.open(r"C:\Users\Administrator\Desktop\china.jpg"))
mask_pic = np.array(Image.open(r"china.jpg"))

wordcloud = WordCloud(font_path=r"C:\\WINDOWS\\Fonts\\simsun.ttc",background_color='black', collocations=False, 
                         max_words=120, max_font_size=150, mask=mask_pic).generate(words)
image=wordcloud.to_image()
image.show()
```
